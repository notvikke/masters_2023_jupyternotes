### Some `lambda` practice

We will use the `lambda` function to sort the list of dictionaries by the `endTime` key.

`lambda` functions are useful when we need to pass a function as an argument to another function. For example, the `sorted` function takes a function as an argument, and uses it to sort the list. These functions are anonymous, so they don't have a name. 

We can use `lambda` functions to create a function on the fly, without having to define it with the `def` keyword.

It's syntax is as follows:

```Python
lambda argument1, argument2, ... : expression
```
This would be the same as:

```Python
def function(argument1, argument2, ...):
    return expression
```

Just as `sorted` takes a function as an argument, `map` takes a function as an argument. `map` applies the function to each element of the list, and returns a new list with the results.

```Python
map(function, list)
```

Let's use `map` to create a new list with the `endTime` key of each dictionary in the list.

````python
map(lambda x: x["endTime"].split()[1][:2], spotify)
````

Remember the lazy evaluation? `map` doesn't return a list, but a `map` object. We can convert it into a list with the `list` function.

This is in order to be efficient with memory. If we had a list with millions of elements, we wouldn't want to create a new list with all the elements in memory. We would want to create the new list when we need it, and not before.

````python
list(map(lambda x: x["endTime"].split()[1][:2], spotify))
````
