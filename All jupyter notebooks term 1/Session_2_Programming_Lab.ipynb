````python
def dividers(num):
    number = 1
    count=0
    while number <= num:
        if num % number == 0:
            count=count+1
            print(" ", num, " Has ", count, "Divisors \n")
        number += 1
        
n=999
j = 1
k = 1
for i in range(1,n):
    print(k)
    dividers(k)
    j = j + 1 
    k = k + j
    
````
