# **S06: INTRODUCTION TO `scikit-learn`**

`scikit-learn` is the most used library for **classic machine learning** in Python. Is a very versatile, flexible and powerful library with a lot of efficient algorithm implementations to solve almost any ML problem. Here is the user guide: https://scikit-learn.org/stable/user_guide.html#user-guide

## **`scikit-learn` basics**

Most commonly, the steps in using the `scikit-Learn` estimator API are as follows.

1. **Choose a model by importing the appropriate estimator from `scikit-Learn`.**
2. **Instantiate the model and choose model hyperparameters**
3. **Arrange data into a features matrix (X) and target vector (y)**
4. **Train the model to your data by calling the `fit()` function**
5. **Apply the model to new data by calling the `predict` function**

We will now step through several simple examples of applying supervised and unsupervised learning methods.

### Supervised learning example: Simple linear regression

Let's read some sample data from `data` folder

```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv("data/sample_data_lr.csv")
```

```python
px.scatter(x=data.X1, y=data.Target, template="none")
```

#### 1. Choose a model by importing the appropiate estimator

In `scikit-Learn`, every model must be imported from the library to be used. So, for example, if we would like to compute a simple linear regression model, we can import the linear regression class

```python
from sklearn.linear_model import LinearRegression
```

Note that other more general linear regression models exist as well; you can read more about them in the [``sklearn.linear_model`` module documentation](http://Scikit-Learn.org/stable/modules/linear_model.html).

#### 2. Instantiate the model, and choose the model hyperparameters

Once we have decided on our model class, there are still some options open to us.
Depending on the model we are working with, we might need to answer one or more questions like the following:

- Would we like to fit for the offset (i.e., *y*-intercept)?
- Would we like the model to be normalized?
- Would we like to preprocess our features to add model flexibility?
- What degree of regularization would we like to use in our model?

These are examples of the important choices that must be made *once the model class is selected*.
These choices are often represented as *hyperparameters*, or parameters that must be set before the model is fit to data.
In Scikit-Learn, hyperparameters are chosen by passing values at model instantiation.

For our linear regression example, we can instantiate the ``LinearRegression`` class and specify that we would like to fit the intercept using the ``fit_intercept`` hyperparameter:

```python
model = LinearRegression(fit_intercept=True)
```

```python
model
```

Keep in mind that when the model is instantiated, the only action is the storing of these hyperparameter values.
In particular, we have not yet applied the model to any data: the Scikit-Learn API makes very clear the distinction between *choice of model* and *application of model to data*.

#### 3. Arrange data into a features matrix and target vector

Inthis example, we already have data in the correct form: `X` and `y`

```python
X = data[["X1"]]
```

```python
X.head()
```

```python
y = data["Target"]
```

```python
y.head()
```

#### 4. Train the model to your data by applying the `fit` function

Now it is time to apply our model to data.
This can be done with the ``fit()`` method of the model:

```python
model.fit(X, y)
```

This ``fit()`` command causes a number of model-dependent internal computations to take place, and the results of these computations are stored in model-specific attributes that the user can explore.
In Scikit-Learn, by convention all model parameters that were learned during the ``fit()`` process have trailing underscores; for example in this linear model, we have the following:

```python
model.coef_
```

```python
model.intercept_
```

#### 5. Apply the model to new data by calling the `predict` function

Once the model is trained, the main task of supervised machine learning is to evaluate it based on what it says about new data that was not part of the training set.
In Scikit-Learn, this can be done using the ``predict()`` method.
For the sake of this example, our "new data" will be a grid of *x* values, and we will ask what *y* values the model predicts:

```python
# fist, we load our "new data"
X_new = pd.read_csv("data/sample_data_lr_new.csv")
```

### Synthetic dataset generation

We also can create our own datasets that follow a specific structure for testing purposes

 - `datasets.make_blobs` - Generate isotropic Gaussian blobs (for clustering)
 - `datasets.make_classification` - Generate a random n-class classification problem (for classfication)
 - `datasets.make_regression` - Generate a random regression problem
 - many more...

```python
from sklearn.datasets import make_blobs, make_classification, make_regression
```

```python
X,y = make_classification(
    n_samples=1000,
    n_features=10,
    n_classes=2
)
```

```python
X = pd.DataFrame(X, columns=[f"X{i}" for i in range(1,11)])
y = pd.Series(y, name="Target")
```

```python
X.head()
```

```python
y.head()
```

### Train/Test split

Helper function to split data for training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,               # arrays or matrices I want to split
    test_size=0.3,      # the proportion to data for testing
    random_state=123,   # can be any number. make the split reproducibile
    shuffle=True,       # if we want to shuffle data before splitting
    stratify=None       # For clasification problems. Split data stratifying the target variable
)
```

Two interesting new concepts here
 - **Reproducibilty:** The execution is reproducible if I can reproduce exactly the same operations with the same results indefinitely. The `random_state` param allows any random operation to output the same results. That is, if the case of `train_test_split`, if we maintain always the same `random_state`, the shuffling operation will be always executes with the same outcome (data will be shuffled exactly the same always)
 - **Stratification:** Sometimes, in classification problems the amount of elements of each class is not proportional (ex: we have 90% of 0s and 10% of 1s). With the `stratify` param we are telling the splitter that we want the same original proportion of 0s and 1s in the resulting parts (train and test).

```python
print(f"X_train size: {X_train.shape}")
print(f"y_train size: {y_train.shape}")
print(f"X_test size: {X_test.shape}")
print(f"X_test size: {y_test.shape}")
```

### Scaling functions

Sometimes, data needs to be previously scaled before fitting any algorithm.
 - `preprocessing.StandardScaler` - Standardize features by removing the mean and scaling to unit variance.
 - `preprocessing.MinMaxScaler` - Transform features by scaling each feature to a given range
 - `preprocessing.MaxAbsScaler` - Scale each feature by its maximum absolute value
 - `preprocessing.RobustScaler` - Removes the median and scales the data according to the quantile range

```python
from sklearn.preprocessing import StandardScaler
```

```python
scl = StandardScaler()
```

```python
scl.fit(X_train)
X_train_sc = scl.transform(X_train)

# the result is returned as a numpy array, so let's convert it again to a Padnas Df
X_train_sc = pd.DataFrame(X_train_sc, columns=X_train.columns)
```

**NOTE 1:** Instead of `fit` and the `transform`, you can use the `fit_transform` method to do both thing sequentially

```python
X_train.describe()
```

```python
X_train_sc.describe()
```

<div class="alert alert-warning">
Be Careful! If you apply scaling transformations to train data, don't forget to apply (`transform`) the same transformation to the test data!
