````python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
````

````python
data = pd.read_csv("data/adult.csv.zip")
data.head()
````

````python
X = data.drop(columns=["income"])
y = data["income"]
````

````python
X.head()
````

````python
y.head()
````

````python
#splitting
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=99)
````

````python
X_train.shape
````

````python
y_train.shape
````

````python
X_test.shape
````

````python
#preprocessing
X_train_cat = X_train.select_dtypes("O")
````

````python
X_train_cat.head()
````

````python
ohe = OneHotEncoder(sparse=False)
````

````python
cat_data_ohe = ohe.fit_transform(X_train_cat)
````

````python
cat_data_ohe
````

````python
cat_data_ohe = pd.DataFrame(cat_data_ohe, columns=ohe.get_feature_names_out())
````

````python
cat_data_ohe.head()
````

````python
X_train_full = pd.concat([X_train.reset_index(), cat_data_ohe], axis=1)
````

````python
X_train_full = X_train_full.drop(columns=X_train_cat.columns)
````

````python
X_train_full.head()
````

````python
lr = LogisticRegression()
````

````python
lr.fit(X_train_full, y_train)
````

````python
X_test_cat = X_test.select_dtypes("O")
X_test_cat.head()
````

````python
X_test_ohe = ohe.transform(X_test_cat)
````

````python
X_test_ohe = pd.DataFrame(X_test_ohe, columns=ohe.get_feature_names_out())
````

````python
X_test_full = pd.concat([X_test.reset_index(), X_test_ohe], axis=1)
````

````python
X_test_full = X_test_full.drop(columns=X_test_cat.columns)
````

````python
lr.predict(X_test_full)
````
