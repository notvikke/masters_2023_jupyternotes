### 2.1 Build the Streaming DataFrame

Let's create the **streaming** DataFrame.

Creating a streaming DataFrame works similarly to the way we a batch DataFrame; we need to use the **readStream** property instead.

In this lab we'll use Kafka as the data source, therefore the following will be the way to create a streaming DataFrame on top:

````python
events_df = (spark.readStream
                 .format("kafka")
                 .option("kafka.bootstrap.servers", "localhost:9092")
                 .option("subscribe", "geo_event")
                 .option("includeHeaders", "true")
                 .option("startingOffsets", "latest")
                 .load())
````

It's just a Spark DataFrame

````python
type(events_df)
````

But a **streaming** one

````python
events_df.isStreaming
````

The dataframe schema is inherit from Kafka messages.

<a id='2.2'></a>
### 2.2 Shape the DataFrame to keep the fields we need

Since *events_df* is a DataFrame, we can apply most of the transformations we already know.

Let's apply some transformations to **calculate top five drivers making driving violations**, in order to know who is not behaving as expected and which kind of violations they're making. That means we'll need the following:

- when the event happens (event time)
- the name of the driver
- the type of violation

The *events_df* DataFrame has only one column of type string named *value*.

We'll consider the structure of this value to **extract the relevant fields we need to implement this use case**:

<img src="https://github.com/georgevetticaden/sam-trucking-data-utils/raw/master/readme-design-artifacts/truck-geo-event.png" />


According to the event data fields:

- event time  is in position 0
- driver name is in position 4
- event type  is in position 7

````python
from pyspark.sql.functions import split, col
from pyspark.sql.types import TimestampType

df = (events_df.select(split("value",'\|').alias("fields"))
               .withColumn("eventTime",col("fields").getItem(0).cast(TimestampType()))
               .withColumn("driverName",col("fields").getItem(4))
               .withColumn("violation", col("fields").getItem(7))
               .select("eventTime", "driverName", "violation"))
````

````python
df
````

<a id='2.3'></a>
### 2.3 Keep just event violations

We are only interested in events related with bad driving practices.
Let's filter the DataFrames so that we only keep actually violations:

````python
# violations would be any of these event types
violations =  ['Unsafe tail distance', 'Overspeed', 'Unsafe following distance']

df = df.where(col("violation").isin(violations))
````

Notice we are reassigning the filtered resulting DataFrame.

````python
df
````

<a id='2.4'></a>
### 2.4 We don't want to consider events that are late more than 5 minutes

In our business case we only care what's happening right now, therefore **5 minutes** it's too late for us and we don't want to consider events arriving so late:

````python
df = df.withWatermark("eventTime", "5 minutes")
````

Notice we are reassigning the filtered resulting DataFrame

````python
df
````

<a id='2.5'></a>
### 2.5 Let's apply a windowed aggreation

We're going to create a sliding window of 1 minute long sliding every 10 seconds.

By using the window function, Spark is going to take care of creating the window boundaries.

````python
from pyspark.sql.functions import window

df = (df.groupBy(
           window("eventTime", "1 minute", "10 seconds"), 
           "driverName",
           "violation"
        )
        .count()
        .orderBy(col("count").desc())
        .limit(5)
     )
````

Notice we are reassigning the filtered resulting DataFrame

````python
df
````

<a id='2.6'></a>
### 2.6 Test the real-time analytics job

Before writing the results to Kafka we can check we get the expected results.

Let's write the results to the console

````python
stream = (df.writeStream
            .format("console")
            .outputMode("complete")
            .queryName("test")
         )

query = stream.start()
query.awaitTermination(30)
query.stop()
````

<a id='2.7'></a>
### 2.7 Let's write the results back into Kafka

In this scenario, Spark is expecting a Dataframe with a least a column called **value**. More about this here: [Structured Streaming + Kafka Integration Guide (Kafka broker version 0.10.0 or higher)](https://spark.apache.org/docs/latest/structured-streaming-kafka-integration.html)

As we're not using any Schema Registry with Kafka, we are going to write back the violations events as strings. 
We're going to use string manipulation functions available in Spark; in particular, we're going to use **concat** to create a string column containing all the information

````python
from pyspark.sql.functions import concat,concat_ws,lit

df = (df.select(concat(col("window.start"),
                       lit("|"),
                       col("window.end"),
                       lit("|"),
                       col("driverName"),
                       lit("|"),
                       col("violation"),
                       lit("|"),
                       col("count")
                      ).alias("value")
               )
     )
````

Notice we are reassigning the filtered resulting DataFrame

````python
df
````

<a id='2.8'></a>
### 2.8 Time to run our real-time analytics job

Now we just need to configure the DataFrame sink, in our case it will be Kafka.

Notice that we did not include any trigger. If no trigger setting is explicitly specified, then by default, the query will be executed in micro-batch mode, where micro-batches will be generated as soon as the previous micro-batch has completed its processing.

````python
stream = (df.writeStream
            .format("kafka")
            .outputMode("complete")
            .option("kafka.bootstrap.servers", "localhost:9092")
            .option("checkpointLocation", "/tmp/spark_checkpoint")
            .option("topic", "top5violations")
            .queryName('top5violations')
         )
````

````python
stream
````

We're ready to start our real-time analytics job.
The last step is to start processing the stream

````python
query = stream.start()
````

we're going to run our job for only 2 minutes (120 seconds) and it'll finish automatically. 

In real life it will run all the time (we won't setup any time in the awaitTermination)

````python
duration = 120

query.awaitTermination(duration)
````

While the app is running, let's read the resulting events written in top5violations topic.

Go back to the console and type the following command:

```sh
kafka-console-consumer.sh --topic top5violations --bootstrap-server localhost:9092
```
