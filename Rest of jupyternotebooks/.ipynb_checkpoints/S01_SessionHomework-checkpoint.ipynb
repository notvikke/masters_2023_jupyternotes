## Exercise 01

Load the following data and answer the questions
 - What is the mean horsepower per brand (you have to extract the brand from the "car" column)
 - Get the heaviest vehicle and the lightest vehicle
 - Plot a boxplot of the displacement, per brand
 - Represent a scatter plot of the displacement and the weight. Are there any relationship?
 - Does the acceleration has anything to do with the mpg?

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("data/auto-mpg.csv")
data.info()
```

```python
data['horsepower'] = data['horsepower'].replace(['?'], np.nan)
data['horsepower'] = pd.to_numeric(data['horsepower'])
data['brand']=data['car'].str.split().str[0]
mean_hp_by_brand = data.groupby('brand')['horsepower'].mean()

mean_hp_by_brand
```

```python

heavy = data[data['weight'] == data['weight'].max()]
print(heavy['car'])
light = data[data['weight'] == data['weight'].min()]
print(light['car'])
```

```python

data.boxplot(by ='brand', column =['displacement'], grid = False)
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['displacement'], data['weight'])
plt.show()
```

```python
plt.scatter(data['acceleration'], data['mpg'])
plt.show()
```

## Exercise 02

Load the GOOG stock price history with `yfinance` library since 01-Jan-2015 

 1. Create a new DataFrame with weekly averaged values of `Close`
 1. Add a new column called `Close_MA` with the weekly moving average of the previous 4 weeks (current week included)
 1. Add a new column called `Close_Std` with the average standard deviation of the previous 4 weeks (current week included) 
 1. Add a new column called `Target` with the shifted `Close` price 1 week. That is, for the week N, the `Target` column should contain the average price corresponding to the N+1 week.
 1. Add a new column called `Target_Bin` with the following binary logic:
    - 0 if the next week's price was lower o equal
    - 1 if the next week's price was higher

```python
import yfinance as yf
goog = yf.Ticker("GOOG")
df = goog.history(start='2015-01-01')
df
```

```python

df['Week'] = df['Date'].dt.week
mean_close_week = df.groupby('Week')['Close'].mean()
mean_close_week =pd.DataFrame(mean_close_week)
mean_close_week 
```

```python
mean_close_week['Close_MA'] = mean_close_week ['Close'].rolling(5).mean()
mean_close_week 
```

```python
mean_close_week ['Close_Std'] = mean_close_week['Close'].rolling(5).std()
mean_close_week 
```

```python
mean_close_week['Target'] = mean_close_week ['Close'].shift(-1)
mean_close_week 
```

```python
mean_close_week['Target_Bin'] = np.where(mean_close_week['Close'] < mean_close_week['Target'], 1, 0)
mean_close_week
```
