```python
df1["Units"] = df1["Units"]+1
```

```python
df1['year-month-week'] = pd.to_datetime(df1['year-week'] + '-1', format='%Y-%W-%w')

# Extract year and month from 'YearWeek' column and create 'YearMonth' column
df1['year-month'] = df1['year-month-week'].dt.strftime('%Y-%m')
```

```python
#df1 = df1.drop(columns="year-week")
```

```python
# Convert YearMonth column to datetime format
df1['year-month'] = pd.to_datetime(df1['year-month'], format='%Y-%m')

# Add a new column with the month name
df1['MonthName'] = df1['year-month'].dt.strftime('%B')
```

```python
df1["year-month"] = df1["year-month"].astype(str)
df1["Year"] = df1["year-month"].str[:4]
```

```python
import math
# define a function that returns the week of the month
def week_of_month(date):
  return math.ceil(date.day / 7)

# apply the function to the date column and create a new column
df1['week_of_month'] = df1['year-month-week'].apply(week_of_month)
```

```python
# Define mapping of months to seasons
season_mapping = {
    'January': 'Winter',
    'February': 'Winter',
    'March': 'Spring',
    'April': 'Spring',
    'May': 'Spring',
    'June': 'Summer',
    'July': 'Summer',
    'August': 'Summer',
    'September': 'Autumn',
    'October': 'Autumn',
    'November': 'Autumn',
    'December': 'Winter'
}

# Add a new column with the season based on the month
df1['Season'] = df1['MonthName'].map(season_mapping)
```

```python
df1 = df1.drop(columns="year-week")
df1 = df1.drop(columns="year-month-week")
df1 = df1.drop(columns="year-month")
```

```python
df1 = pd.get_dummies(df1, columns=['MonthName', 'Year','week_of_month','Season'])
```

```python
X = df1.drop(columns="Units")
y = df1["Units"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_percentage_error
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=8, random_state=42,shuffle=False)
```

```python
import pandas  as pd
import numpy   as np
import xgboost as xgb

regressor=xgb.XGBRegressor(**best_para)
regressor.fit(X_train, y_train)
predictions = regressor.predict(X_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, predictions))
print(f"RMSE: {rmse}")
```

```python
plt.plot(y_test.index, y_test, label='Actual')
plt.plot(y_test.index, predictions, label='Predicted')
plt.xlabel('Date')
plt.ylabel('Units')
plt.legend()
plt.show()

from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test, predictions)
```
